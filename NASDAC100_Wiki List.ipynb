```python
import requests
from bs4 import BeautifulSoup

def Get_Neighborhood(url):
    res = requests.get(url)
    #The names in the first ul tag
    html_parsed = BeautifulSoup(res.text, 'html.parser').find('ol')
    html = BeautifulSoup(str(html_parsed), 'html.parser')
    out = []
    for x in html.find_all('li'):
        if x.string != None:
            out.append(x.string)
    return (out)
Neighborhoods = Get_Neighborhood('https://en.wikipedia.org/wiki/NASDAQ_Financial-100')

for x in Neighborhoods:
    print (x)
```
